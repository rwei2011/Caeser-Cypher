1. Write a function cumulative_sum that accepts an integer parameter n and returns the sum of the numbers 1 through n, including n.
For example:
Call                Returns
cumulative_sum(1)   1
cumulative_sum(5)   15


```python
#Iterative Solution
def cumulative_sum(n):
  counter = 0
  for i in range(1,n+1):
    counter += i
  print(counter)
cumulative_sum(5)
#Recursive
def csum_rec(n):
  if n == 1:
    return 1
  else:
    return n + csum_rec(n-1)
print(csum_rec(5))
```

2. Write a function count_item that accepts two parameters: a list of integers values and another integer x. Return the number of times the integer x appears in values.
For example:
Call                                Returns
count_item([], 1)                    0
count_item([1,1,1,1], 1)             4
count_item([1, 2, 3, 1, 2], 1)       2
count_item([1, 2, 3, 1, 2], 5)       0


```python
def count_item(list1,x):
  count = 0
  for i in list1:
    if i == x:
      count += 1
  print(count)
count_item([1,1,1,1], 1)
#recusive
def count_item_rec(lst,x):
  if not lst:
    return 0
  elif lst[0] == x:
    return 1 + count_item_rec(lst[1:],x)
  else:
    return count_item_rec(lst[1:],x)
count_item_rec([1,1,1,1], 1)
'''
call stack
cir([1,2,3],1) = 1 + cir([2,3],1) --> 1
cir([2,3],1) = cir([3],1) --> 0
cir([3],1) = cir([],1) --> 0
cir([],1) = 0
```

3. Write a function count_strings that accepts two parameters: a list of strings words and another integer length. Return the number of strings in words that are exactly length long.
For example:
Call                                                    Returns
count_strings(["Up", "Down", "Left", "Right"], 4)       2
count_strings(["Up", "Down", "Left", "Right"], 2)       1
count_strings(["Hello", "Goodbye"], 5)                  1  


```python
def count_strings(lst,n):
  count = 0
  for i in lst:
    if len(i) == n:
      count += 1
  print(count)
count_strings(["Up", "Down", "Left", "Right"], 2)
def count_strings_rec(lst,n):
  if not lst:
    return 0
  elif len(lst[0]) == n:
    return 1 + count_strings_rec(lst[1:],n)
  else:
    return count_strings_rec(lst[1:],n)
print(count_strings_rec(["Hello", "Goodbye"], 5))
```

4. Write a function star_string that accepts an integer parameter n and returns a string of stars (asterisks) 2^n long (i.e., 2 to the nth power).
For example:
Call            Returns      Reason
star_string(0)  "*"          2^0 = 1
star_string(1)  "**"         2^1 = 2
star_string(2)  "****"       2^2 = 4
star_string(3)  "********"   2^3 = 8


```python
def star_string(n):
  len1 = 2 ** n
  return "*" * len1
print(star_string(3))
def star_string_rec(n):
  if n == 0:
    return "*"
  else:
    return 2 * star_string_rec(n-1)
print(star_string_rec(4))
```
